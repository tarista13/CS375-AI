```python
import numpy as np

y_true = np.array([1.,2.,3.])

y_pred_a = np.array([1.5,1.5,3.5])
y_pred_b = np.array([1.1,2.1,1.8])

#MAE: Mean Absolute Error: average of absolute differences
#|time - A/B| is the sum of errors/mean absolute error(MAE)
mae_a = np.abs(y_true - y_pred_a).mean()
mae_b = np.abs(y_true - y_pred_b).mean()

print('MAE a: {:.2f}, MAE b: {:.2f}'.format(mae_a, mae_b))

#MSE: Mean Squared Error or
#RMSE: Root Mean Squared Error
#(time - A/B)^2 is average/mean error
mse_a = ((y_true - y_pred_a) ** 2).mean()
mse_b = ((y_true - y_pred_b) ** 2).mean()

print('MSE a: {:.2f}, MSE b: {:.2f}'.format(mse_a, mse_b))
```
